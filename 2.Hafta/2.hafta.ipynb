```python
from sklearn.datasets import load_iris

iris = load_iris()
iris
```

```python
import pandas as pd

data = pd.DataFrame(iris.data)

etiket = pd.DataFrame(iris.target)

egitim_tablosu = pd.concat([data,etiket],axis=1)

print(egitim_tablosu.shape)

egitim_tablosu = egitim_tablosu.set_axis(['sepal length', 'sepal width','petal length', 'petal width', 'etiket'],axis=1)

egitim_tablosu
```

```python
egitim_tablosu['etiket'].value_counts()
```

```python
data.head()
```

```python
etiket.head()
```

```python
iris.target_names
```

```python
from sklearn.model_selection import train_test_split

egitim_tablosu,test_tablosu, egitim_etiketi, test_etiketi = train_test_split(data,etiket,test_size=0.20)
```

```python
pd.DataFrame(egitim_tablosu)
```

```python
from sklearn.tree import DecisionTreeClassifier as karar_agaci
model = karar_agaci()

model.fit(egitim_tablosu,egitim_etiketi)
```

```python
modelin_tahminleri = model.predict(test_tablosu)
modelin_tahminleri
```

```python
dogru_tahmin_orani = model.score(test_tablosu,test_etiketi)
print(f"Modelin dogru tahmin etme orani : {dogru_tahmin_orani}")
print(f"Modeli hata payi : {1 - dogru_tahmin_orani}")
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

tree.plot_tree(model,
               feature_names=iris.feature_names,
               class_names=iris.target_names,
               impurity=True
              )
plt.figure(figsize=(40,40),dpi=100)
```

```python
from sklearn import metrics

metrics.confusion_matrix(test_etiketleri,modelin_tahminleri)
```
