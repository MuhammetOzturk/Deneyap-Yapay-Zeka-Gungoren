```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(6,input_dim = 2, activation = 'relu'))
model.add(Dense(12,activation='relu'))
model.add(Dense(12,activation='relu'))
model.add(Dense(12,activation='relu'))
model.add(Dense(12,activation='relu'))
model.add(Dense(12,activation='relu'))
model.add(Dense(12,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy', optimizer='adam')
model.summary()
```

```python
import pandas as pd
tablo = pd.read_csv('INTELLIGENT IRRIGATION SYSTEM.csv')
```

```python
tablo.head()
```

```python
data = tablo[['RH (%)','SICAKLIK']]
data.head()
```

```python
pompa = tablo['POMPA']
pompa.head()
```

```python
from sklearn.model_selection import train_test_split

egitim_tablosu, test_tablosu, egitim_pompa, test_pompa = train_test_split(data,pompa, test_size=0.20)
```

```python
len(egitim_tablosu),len(test_tablosu),len(egitim_pompa),len(test_pompa)
```

```python
model.fit(egitim_tablosu,egitim_pompa,epochs=60,batch_size=5)
```

```python
model.predict(test_tablosu)
```
